### Separate 'Date' into each column

```python
dfReport['Year'] = pd.DatetimeIndex(dfReport['Date']).year
dfReport['Month'] = pd.DatetimeIndex(dfReport['Date']).month
dfReport['Day'] = pd.DatetimeIndex(dfReport['Date']).day
```

### Clean some data
Fatal (Y/N): Whether or not the victim died from the attack. (String)

#### Fatal column

```python
dfReport['Fatal (Y/N)'].unique()
```

There are some invalid values. I will convert 'y' to 'Y' and other than 'N' and 'Y' will be None

```python
for index in dfReport.index:
    value = dfReport.at[index, 'Fatal (Y/N)']
    if value == 'Y' or value == 'N':
        continue
    else:
        if value == 'y':
            dfReport.at[index, 'Fatal (Y/N)'] = 'Y'
        else:
            dfReport.at[index, 'Fatal (Y/N)'] = None
```

```python
dfReport['Fatal (Y/N)'].unique()
```

#### Age column
Convert string to integer and convert all invalid values to None

```python
dfReport['Age'].unique()
```

```python
dfReport['Age'].notna().index
```

```python
dfTemp = dfReport.reset_index()
for index in dfTemp.loc[dfTemp['Age'].notna()].index:
    value = dfTemp.at[index, 'Age']
    original_index = dfTemp.at[index, 'index']
    if value.isdigit():
        dfReport.at[original_index, 'Age'] = int(dfReport.at[original_index, 'Age'])
    else:
        dfReport.at[original_index, 'Age'] = None
```
